```python
#! /usr/bin/env python
# -*- coding: utf-8 -*-

import sys
from ROOT import *
import numpy as np

#Parameters
gamma=2
m=90
events=1e6
event=0
```

```python
#We create the canvas and the histogram

h1 = TH1D('h1','BreitWigner distribution',200, 70., 110)
c1 = TCanvas('c1','c1',1000,500)
```

```python
#We obtain the random breit wigner distribution values

while True:
    #Events generator
    E = gRandom.BreitWigner(m,gamma)
    h1.Fill(E)
    
    #Limiter
    event+=1
    if event==6000:
        break
        
```

```python
#Now we plot
h1.GetXaxis().SetTitle("m (MeV) ")
h1.GetYaxis().SetTitle("counts")
h1.Draw()
c1.Draw()
```
